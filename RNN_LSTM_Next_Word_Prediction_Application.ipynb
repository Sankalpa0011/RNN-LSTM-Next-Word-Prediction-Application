## **Import Libraries**

```python
import numpy as np
import tensorflow as tf
```

## **Open Text Dataset**

```python
with open("/content/1661-0.txt", "r", encoding="utf-8") as file:
  text = file.read()
```

```python
text
```

```python
# filters = '!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t'
```

## **Define A Tokernizer**

```python
tokernizer = tf.keras.preprocessing.text.Tokenizer()
```

```python
tokernizer.fit_on_texts([text])
```

## **Numerical Representation and Word Counts and Other Details**

```python
tokernizer.get_config()
```

```python
# Get the word indexes
tokernizer.word_index
```

```python
input_sequences = []

# Splitting sentenses by using newline character
for sentense in text.split('\n'):
  print(sentense)
  token_list = tokernizer.texts_to_sequences([sentense])[0]  # Convert To Numerical
  print(token_list)
  for i in range(1, len(token_list)):
    sequence = token_list[:i+1]
    input_sequences.append(sequence)
```

```python
input_sequences[:5]
```

## **Add padings to set all the input sequences to same length**

```python
# Get maximum sequence length
max_length = max([len(input_sequences) for input_sequences in input_sequences])
max_length
```

```python
# Add padings
input_sequences = tf.keras.preprocessing.sequence.pad_sequences(input_sequences, maxlen=max_length, padding="pre")
input_sequences[:5]
```

```python
input_sequences[0]
```

```python
len(input_sequences[0])
```

## **Feature and label Extraction**

```python
x = input_sequences[:, :-1]  # Feature
y = input_sequences[:, -1]   # Label
```

```python
# Fetures of 1st record
x[0]
```

```python
# label of the 1st record
y[0]
```

## **One Hot Encoding**

```python
num_classes = len(tokernizer.word_index) + 1
num_classes
```

```python
# Encoding
y = np.array(tf.keras.utils.to_categorical(y, num_classes=num_classes))
```

## **Model Building**

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Embedding(num_classes, 100, input_length=max_length-1))
model.add(tf.keras.layers.LSTM(128))
model.add(tf.keras.layers.Dense(num_classes, activation="softmax"))

#compile
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

#summary
model.summary()
```

```python
# Train the model
history = model.fit(x, y, epochs=100)
```
